```python
import numpy as np
import matplotlib.pyplot as plt

from defdap.quat import Quat
from defdap import defaults
from defdap.plotting import PolePlot

%matplotlib inline
```

```python
e1 = np.array([20., 30., 10.]) * np.pi / 180
q1 = Quat.from_euler_angles(*e1)

defaults['ipf_triangle_convention'] = 'down'
q1.plot_ipf([0,0,1], 'hexagonal')
defaults['ipf_triangle_convention'] = 'up'
q1.plot_ipf([0,0,1], 'hexagonal')
```

```python
e1 = np.array([20., 30., 10.]) * np.pi / 180
q1 = Quat.from_euler_angles(*e1)

sym = 'hexagonal'
# sym = 'cubic'
defaults['ipf_triangle_convention'] = 'down'
fig, axes = plt.subplots(1, 3, figsize=(10, 4))
q1.plot_ipf([1,0,0], sym, ax=axes[0])
q1.plot_ipf([0,1,0], sym, ax=axes[1])
q1.plot_ipf([0,0,1], sym, ax=axes[2])
```

```python
plot = PolePlot('IPF', 'cubic')

plot.add_line((0, 0, 1), (1, 1, 1), plot_syms=True)
plot.add_line((0, 0, 1), (1, 0, 1), plot_syms=True)
plot.add_line((1, 0, 1), (1, 1, 1), plot_syms=True)

plot.label_point((0,0,1), plot_syms=True)
plot.label_point((1, 1, 1), plot_syms=True)
plot.label_point((1, 0, 1), plot_syms=True)
```

```python
plot = PolePlot('IPF', 'hexagonal')

plot.add_line((0, 0, 0, 1), (-1, 2, -1, 0), plot_syms=True)
plot.add_line((0, 0, 0, 1), (-1, 1, 0, 0), plot_syms=True)
plot.add_line((-1, 2, -1, 0), (-1, 1, 0, 0), plot_syms=True)

plot.label_point((0, 0, 0, 1), plot_syms=True)
plot.label_point((-1, 2, -1, 0), plot_syms=True)
plot.label_point((-1, 1, 0, 0), plot_syms=True)
```

```python
n_oris = 500000
sym = 'cubic'

eulers = np.random.uniform(size=(n_oris, 3)) * [2*np.pi, 1, 2*np.pi]
eulers[:, 1] = np.acos(2*eulers[:, 1]-1)
quats = Quat.create_many_quats(eulers.T)
ipf_colours = Quat.calc_ipf_colours(quats, [0,0,1], sym)

Quat.plot_ipf(quats, [0,0,1], sym, marker=None, color=ipf_colours.T, s=0.5)
```
